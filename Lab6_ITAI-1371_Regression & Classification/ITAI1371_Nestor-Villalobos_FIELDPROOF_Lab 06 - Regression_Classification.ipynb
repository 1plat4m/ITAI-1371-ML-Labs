## Introduction
This lab applies regression and classification models to the FIELDPROOF dataset to predict compliance and analyze relationships between variables.

```python
import pandas as pd
import numpy as np
import requests
from io import StringIO
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
url = "https://digitallycreative.net/data-fieldproof/FIELDPROOF_simulated_dataset.csv"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers, timeout=30)
response.raise_for_status()
df = pd.read_csv(StringIO(response.text))

# Features available in the FIELDPROOF dataset.
feature_cols = ["duration_min", "motion_intensity", "posture_correctness", "zone_presence_score", "sequence_score", "fatigue_index", "anomaly_score", "risk_score"]
df_model = df[feature_cols].copy()

# Regression target: continuous compliance score.
df_model["compliance_score"] = df["compliance_score"]

# Classification target: Verified vs non-Verified.
df_model["verified"] = (df["verification_status"] == "Verified").astype(int)

df_model.head()
```

```python
X = df_model[feature_cols]
y_reg = df_model["compliance_score"]
y_class = df_model["verified"]

X_train, X_test, y_train_reg, y_test_reg, y_train_class, y_test_class = train_test_split(
    X, y_reg, y_class, test_size=0.3, random_state=42
)
```

## Regression Model

```python
reg = LinearRegression()
reg.fit(X_train, y_train_reg)

y_pred_reg = reg.predict(X_test)

print("MSE:", mean_squared_error(y_test_reg, y_pred_reg))
print("R2:", r2_score(y_test_reg, y_pred_reg))
```

## Classification Model

```python
clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train_class)

y_pred_class = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test_class, y_pred_class))
```

## Analysis
Regression predicts numeric compliance trends while classification predicts categories.
