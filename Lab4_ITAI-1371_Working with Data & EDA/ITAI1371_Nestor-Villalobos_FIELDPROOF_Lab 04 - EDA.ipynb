## Introduction

This lab focuses on performing Exploratory Data Analysis (EDA) using a simulated FIELDPROOF dataset.

FIELDPROOF is a system designed to verify real-world human activity using biometric and sensor data.

Goals:
- Understand the dataset
- Identify patterns
- Detect trends
- Prepare for ML modeling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "heart_rate": [72,85,90,60,78,88,95,70,76,82],
    "body_temp": [98.6,99.1,100.2,97.9,98.7,99.5,100.5,98.0,98.4,99.0],
    "movement_score": [80,65,50,90,85,55,45,88,82,70],
    "work_duration": [6,5,4,8,7,4,3,7,6,5],
    "compliance": [1,1,0,1,1,0,0,1,1,1]
}
df = pd.DataFrame(data)
df
```

```python
df.head()
df.info()
df.describe()
```

## Dataset Overview

The dataset includes biometric signals and activity data.

- Compliance = 1 (success), 0 (failure)
- Values fall within normal human ranges

```python
df.hist(figsize=(10,6))
plt.show()
```

## Data Distribution Analysis

- Heart rate and temperature are realistic
- Movement varies widely
- Dataset is suitable for modeling

```python
df.corr()
```

```python
plt.imshow(df.corr(), cmap='viridis')
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation=45)
plt.yticks(range(len(df.columns)), df.columns)
plt.title("Correlation Heatmap")
plt.show()
```

## Correlation Insights

- Movement score correlates with compliance
- Work duration impacts performance

```python
df.groupby('compliance').mean()
```

## Compliance Insights

- Higher activity = higher compliance
- Lower movement often = non-compliance
