## Introduction

This lab focuses on preparing data for machine learning using the FIELDPROOF dataset. Data preparation includes cleaning, handling missing values, encoding, and scaling.

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```
data = {
    "heart_rate": [72,85,90,np.nan,78,88,95,70,76,82],
    "body_temp": [98.6,99.1,100.2,97.9,np.nan,99.5,100.5,98.0,98.4,99.0],
    "movement_score": [80,65,50,90,85,55,45,88,82,70],
    "work_duration": [6,5,4,8,7,4,3,7,6,5],
    "compliance": [1,1,0,1,1,0,0,1,1,1]
}
df = pd.DataFrame(data)
df
```

## Step 1: Handle Missing Values

```
df.isnull().sum()
```

```
df.fillna(df.mean(), inplace=True)
df
```

### Explanation
Missing values were replaced using the mean of each column to maintain dataset consistency.

## Step 2: Feature Selection

```
X = df.drop('compliance', axis=1)
y = df['compliance']
```

### Explanation
Features (X) are the input variables, and compliance is the target variable.

## Step 3: Feature Scaling

```
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Explanation
Scaling ensures all features have equal importance and improves model performance.

## Step 4: Final Dataset

```
pd.DataFrame(X_scaled, columns=X.columns)
```
